## Concatenate & clean up

```python
frames = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15]
```

```python
#concatenate dataframes
df = pandas.concat(frames, sort=False)
```

```python
df.columns
```

```python
df_cleaned = df[["cell","overall_Number of packets", "overall_Data byte rate", "overall_Data bit rate", "overall_Average packet size", "overall_Average packet rate", "incoming_Number of packets", "incoming_Data byte rate", "incoming_Data bit rate", "incoming_Average packet size", "incoming_Average packet rate", "outgoing_Number of packets", "outgoing_Data byte rate", "outgoing_Data bit rate", "outgoing_Average packet size", "outgoing_Average packet rate"]]
```

```python
df_cleaned = df_cleaned.astype(str)
```

```python
df_cleaned["overall_number_of_packets"] = df_cleaned["overall_Number of packets"].astype("int64")
```

```python
df_cleaned.drop('overall_Number of packets', axis=1, inplace=True)
```

```python
df_cleaned["incoming_number_of_packets"] = df_cleaned["incoming_Number of packets"].astype("int64")
```

```python
df_cleaned.drop('outgoing_Number of packets', axis=1, inplace=True)
```
